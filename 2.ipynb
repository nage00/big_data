### 2. 유럽국가들에 대해서는 맥주와 와인간의 상관관계는 얼마나 되는지 피어슨상관값(Pearson Correlation)을 구하기. (유럽국가들에 대해서만 맥주와 와인의 데이터를 뽑은 후에 피어슨상관값을 구하는 함수를 적용해야 함)- 02-drinks-eda.ipynb파일을 이용해야 함.

```python
import pandas as pd
import numpy as np
```

#### drinks 데이터를 받아옴. 

```python
file_path = '../data/drinks.csv'
drinks = pd.read_csv(file_path)
```

```python
drinks.head(10)
```

#### drinks 데이터 중 유럽(EU) 데이터만 가져옴. 

```python
europe = drinks.loc[drinks['continent']=='EU']
```

```python
europe
```

#### 유럽국가들에 대한 맥주와 와인간의 피어슨상관값(Pearson Correlation)을 구함. 

```python
corr = europe[['beer_servings', 'wine_servings']].corr(method = 'pearson')
corr
```

#### 유럽국가들에서 맥주와 와인간의 상관관계는 약 30% 로, 미약한 상관관계가 있음.
