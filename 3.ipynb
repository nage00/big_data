### 3. 한화팀과 SK팀간의 2017년 연봉의 평균간에는 유의미한 차이가 존재하는지 t-test로 검정하기. t-test을 사용하는 예는 02-drinks-eda.ipynb파일에 있고, 야구팀의 데이터는 01-baseball-salary-prediction.ipynb파일에 있음.

```python
import pandas as pd
import numpy as np
from scipy import stats
```

#### 야구팀 데이터를 받아옴. (picher; 투수 batter; 타자)

```python
picher_file_path = '../data/picher_stats_2017.csv'
batter_file_path = '../data/batter_stats_2017.csv'
picher = pd.read_csv(picher_file_path)
batter = pd.read_csv(batter_file_path)
```

```python
picher.head()
```

```python
batter.head()
```

```python
picher.groupby('팀명').describe()
```

#### 한화팀과 SK팀의 투수의 2017년 연봉의 평균 간에는 유의미한 차이가 존재하는지 t-test로 검정

```python
hw_p = picher.loc[picher['팀명']=='한화']['연봉(2017)']
sk_p = picher.loc[picher['팀명']=='SK']['연봉(2017)']
```

```python
hw_p.mean()
```

```python
sk_p.mean()
```

```python
tTestResult = stats.ttest_ind(hw_p, sk_p)
tTestResultDiffVar = stats.ttest_ind(hw_p, sk_p, equal_var=False)

print("The t-statistic and p-value assuming equal variances is %.3f and %.3f." % tTestResult)
print("The t-statistic and p-value not assuming equal variances is %.3f and %.3f" % tTestResultDiffVar)
```

p-value 값이 0.05보다 크다. 
그러므로 이 차이는 우연히 발생한 차이로, 인과관계가 존재하지 않음. 

→ 한화팀과 SK팀의 투수의 2017년 연봉의 평균 간에는 유의미한 차이가 존재하지 않음. 

#### 한화팀과 SK팀의 타자의 2017년 연봉의 평균 간에는 유의미한 차이가 존재하는지 t-test로 검정

```python
hw_b = batter.loc[batter['팀명']=='한화']['연봉(2017)']
sk_b = batter.loc[batter['팀명']=='SK']['연봉(2017)']
```

```python
tTestResult = stats.ttest_ind(hw_b, sk_b)
tTestResultDiffVar = stats.ttest_ind(hw_b, sk_b, equal_var=False)

print("The t-statistic and p-value assuming equal variances is %.3f and %.3f." % tTestResult)
print("The t-statistic and p-value not assuming equal variances is %.3f and %.3f" % tTestResultDiffVar)
```

```python
p-value 값이 0.05보다 크다. 
그러므로 이 차이는 우연히 발생한 차이로, 인과관계가 존재하지 않음. 

→ 한화팀과 SK팀의 타자의 2017년 연봉의 평균 간에는 유의미한 차이가 존재하지 않음. 
```
